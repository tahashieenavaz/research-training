```python
from torchvision.models import resnet50
from torchvision.datasets import CIFAR10
from torchvision import transforms
from copy import deepcopy
import numpy as np
import torch
import matplotlib.pyplot as plt
import random
```

```python
class Settings:
  # Number of samples chosen uniformly at random from validation (test) set
  # to purturb the model further on. Originally, it was set to 128 in all experiments (K=128).
  random_samples = 128
  assert random_samples == 128

  # Number of models in the final ensemble (S = [15, 30])
  ensemble_count = 30
  assert ensemble_count in [15, 30]
```

```python
base_model = resnet50(pretrained=True)
```

```python
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        (0.5, 0.5, 0.5),
        (0.5, 0.5, 0.5),
    )
])

test_dataset = CIFAR10(
    root="./data",
    train=False,
    download=True,
    transform=transform
)
```

```python
image = test_dataset[100][0]
image = image / 2 + 0.5
image = np.transpose(image, (1, 2, 0))

plt.imshow(image)
plt.show()
```

```python
models = []

for _ in range(Settings.ensemble_count):
  copied_model = deepcopy(base_model)

  criterion = torch.nn.CrossEntropyLoss()
  optimizer = torch.optim.Adam(copied_model.parameters(), lr=0.001)

  indices = random.sample(range(len(test_dataset)), Settings.random_samples)
  batch   = torch.utils.data.Subset(test_dataset, indices)
  batch   = torch.utils.data.DataLoader(batch, batch_size=Settings.random_samples, shuffle=False)

  images, labels = next(iter(batch))

  copied_model.train()
  optimizer.zero_grad()
  predictions = copied_model(images)
  loss = criterion(predictions, labels)
  print(loss)
  loss.backward()
  optimizer.step()

  models.append(copied_model)
```
